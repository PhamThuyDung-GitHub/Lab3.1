```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x, y, 'r') # 'r' is the color red
plt.xlabel('X Axis title here')
plt.ylabel('Y Axis title here')
plt.title('String title here')
plt.show()
```

```python
# plt.subplot(nrows, ncols, plot(number))
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r--') # More on color options
plt.subplot(1, 2, 2)
plt.plot(y, x, 'g*-');
```

```python
# Create figure (empty canvas)
fig = plt.figure()

# Add set of axes to figure
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, heihgt (range of the model)

# Plot on that set of axes
axes.plot(x, y, 'b')
axes.set_xlabel('Set X label') # Notice the user of set to begin method
axes.set_ylabel('Set y label')
axes.set_title('Set title')
```

```python

# Creating a blank canvas
fig = plt.figure()

# Adding main axes and inset axes to the canvas
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

# Plotting on main axes and setting its labels and title
axes1.plot(x, y, 'b')
axes1.set_xlabel('X_label_axes1')
axes1.set_ylabel('Y_label_axes1')
axes1.set_title('Axes 1 Title')

# Plotting on inset axes and setting its labels and title
axes2.plot(x, y, 'r')
axes2.set_xlabel('X_label_axes2')
axes2.set_ylabel('Y_label_axes2')
axes2.set_title('Axes 2 Title')

# Displaying the plot
plt.show()
```

```python
# Use similar to plt.figure() except use tuple unpacking to grab fig and axes
import matplotlib.pyplot as plt

fig, axes = plt.subplots()

axes.plot(x, y, 'r')
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')

plt.show()
 
```

```python
# Empty canvas of 1 by 2 subplots
fig, axes = plt.subplots(nrows = 1, ncols = 2)
```

```python
# Axes is an array of axes to plot on
axes
```

```python

fig, axes = plt.subplots(nrows=1, ncols=2)

# Plot data on each axes
for ax in axes:
    ax.plot(x, y, 'b')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')

# Display the figure object
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x, y, 'g')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].set_title('title')

axes[1].plot(x, y, 'g')
axes[1].set_xlabel('x')
axes[1].set_ylabel('y')
axes[1].set_title('title')

plt.tight_layout()
plt.show()
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x**2, label = "x**2")
ax.plot(x, x**3, label = "x**3")
ax.legend()
```

```python
fig, axes = plt.subplots(1, 3, figsize = (12, 4))

axes[0].plot(x, x**2, x, x**3)
axes[0].set_title("default axes ranges")

axes[1].plot(x, x**2, x, x**3)
axes[1].axis('tight')
axes[1].set_title("tight axes")

axes[2].plot(x, x**2, x, x**3)
axes[2].set_ylim([0,60])
axes[2].set_xlim([2, 5])
axes[2].set_title("custom axes range");
```
